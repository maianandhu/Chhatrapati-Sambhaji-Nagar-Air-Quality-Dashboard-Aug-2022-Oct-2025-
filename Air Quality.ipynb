```python
df = pd.read_csv("Cleaning.csv",header=3)
df = df.drop(columns=['carbon_dioxide (ppm)']) #droping this cause co2 is not a pollutant
df.dropna(inplace=True)
df.reset_index(drop=True,inplace = True)
```

```python
def time_of_the_day(hour):
    if 5<= hour <12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17<= hour < 19:
        return 'Evening'
    else: 
        return 'Night'
df['time_of_the_day'] = df['hour'].apply(time_of_the_day)
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df
```

```python
df['carbon_monoxide (μg/m³)'] = df['carbon_monoxide (μg/m³)'] / 1000


def aqi_pm25(c):
    if c <= 30: 
        return c * (50/30)
    elif c <= 60: 
        return 50 + (c - 30) * (50/30)
    elif c <= 90: 
        return 100 + (c - 60) * (100/30)
    elif c <= 120: 
        return 200 + (c - 90) * (100/30)
    elif c <= 250: 
        return 300 + (c - 120) * (100/130)
    else: 
        return 400 + (c - 250) * (100/250)

def aqi_pm10(c):
    if c <= 50: 
        return c
    elif c <= 100: 
        return 50 + (c - 50)
    elif c <= 250: 
        return 100 + (c - 100) * (100/150)
    elif c <= 350: 
        return 200 + (c - 250)
    elif c <= 430: 
        return 300 + (c - 350) * (100/80)
    else: 
        return 400 + (c - 430) * (100/70)

def aqi_no2(c):
    if c <= 40: 
        return c * (50/40)
    elif c <= 80: 
        return 50 + (c - 40) * (50/40)
    elif c <= 180: 
        return 100 + (c - 80) * (100/100)
    elif c <= 280: 
        return 200 + (c - 180)
    elif c <= 400: 
        return 300 + (c - 280) * (100/120)
    else: 
        return 400 + (c - 400) * (100/100)

def aqi_so2(c):
    if c <= 40: 
        return c * (50/40)
    elif c <= 80: 
        return 50 + (c - 40) * (50/40)
    elif c <= 380: 
        return 100 + (c - 80) * (100/300)
    elif c <= 800: 
        return 200 + (c - 380) * (100/420)
    elif c <= 1600:
        return 300 + (c - 800) * (100/800)
    else: 
        return 400 + (c - 1600) * (100/400)

def aqi_co(c):
    if c <= 1:
         return c * (50 / 1)
    elif c <= 2:
        return 50 + (c - 1) * (50 / 1)
    elif c <= 10:
        return 100 + (c - 2) * (100 / 8)
    elif c <= 17:
        return 200 + (c - 10) * (100 / 7)
    elif c <= 34:
        return 300 + (c - 17) * (100 / 17)
    else:
        return 400 + (c - 34) * (100 / 17)


def aqi_o3(c):
    if c <= 50: 
        return c
    elif c <= 100: 
        return 50 + (c - 50)
    elif c <= 168: 
        return 100 + (c - 100) * (100/68)
    elif c <= 208: 
        return 200 + (c - 168) * (100/40)
    elif c <= 748: 
        return 300 + (c - 208) * (100/540)
    else: 
        return 400 + (c - 748) * (100/252)
```

```python
df['AQI_PM2_5'] = df['pm2_5 (μg/m³)'].apply(aqi_pm25)
df['AQI_PM10'] = df['pm10 (μg/m³)'].apply(aqi_pm10)
df['AQI_NO2'] = df['nitrogen_dioxide (μg/m³)'].apply(aqi_no2)
df['AQI_SO2'] = df['sulphur_dioxide (μg/m³)'].apply(aqi_so2)
df['AQI_CO'] = df['carbon_monoxide (μg/m³)'].apply(aqi_co)
df['AQI_O3'] = df['ozone (μg/m³)'].apply(aqi_o3)
```

```python
df['AQI'] = df[['AQI_PM2_5', 'AQI_PM10', 'AQI_NO2', 'AQI_SO2', 'AQI_CO', 'AQI_O3']].max(axis=1)
```

```python
df.to_csv("Air Quality Cleaned.csv",index=False)
```
